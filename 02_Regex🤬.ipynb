# Regex : expressions r√©guli√®re

On connsais, tu donne des regles et sous forme d'expressions r√©guli√®re ``r"[a-Z]+@[a-z]+\.com"`` pour trouver une adresse mail par exemple

Test en ligne https://regex101.com

```python
import re 

text = 'Mon adresse mail est blabla@gmail.com'
pattern = r"[a-z]+@[a-z]+\.com"
match = re.search(pattern, text)

print(f'Trouv√© üòÅ: {match.group()}' if match else 'Rien trouv√© üò¢')
```

# Explication 





```python
import re
do_match = [
    'romain@hotmail.com',
    'jeanyves@yahoo.com',
    'romain@hotmail.fr',
    'jean.yves@yahoo.com'
]

dont_match = [
    'jean-yves@yahoo.com',
    'romain@hotmail.gmail.fr',
    'roger@yahoo'
]

regex = r'[a-z\.]+@[a-z]+\.+[a-z]{1,3}\Z'

for email in do_match:
    if re.search(regex, email):
        print(f'it\'s a match : {email}')
    else:
        print(f'no match : {email}')


print('STOOOOOP')
for email in dont_match:
    if re.search(regex, email):
        print(f'it\'s a match : {email}')
    else:
        print(f'no match : {email}')
```
