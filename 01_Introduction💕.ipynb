```python
# New pour r√©cuperer uniquement la premiere occurence
df_fr = train[train.language == 'fr']
review = next(review for _, review in df_fr.review_body.items() if len(review) > 1000)
print(review)
```

#### On va essayer diff√©rent type de tokenisation pour voir les diff√©rence 

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(review)
print(tokens)
print(len(tokens))
```

```python
from nltk.tokenize import wordpunct_tokenize
tokens = wordpunct_tokenize(review)
print(tokens)
print(len(tokens))
```

```python
from nltk.tokenize import sent_tokenize
tokens = sent_tokenize(review)
print(tokens)
print(len(tokens))
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokens = TreebankWordTokenizer().tokenize(review)
print(tokens)
print(len(tokens))
```

```python
from nltk.tokenize import ToktokTokenizer
tokens = ToktokTokenizer().tokenize(review)
print(tokens)
print(len(tokens))
```

```python
from nltk.tokenize import TweetTokenizer
tokens = TweetTokenizer().tokenize(review)
print(tokens)
print(len(tokens))
```

```python
from nltk.tokenize import TweetTokenizer
tokenizer = TweetTokenizer()
vocab = set(
    token for _, review in df_fr['review_body'].items() for token in tokenizer.tokenize(review)
)
print(len(vocab))
print(vocab)
```

```python
from nltk.corpus import stopwords
vocab = vocab - set(stopwords.words('french'))
vocab = vocab - set(stopwords.words('english'))
print(len(vocab))
```

# Stemming :
Le fait de revenir a la taille radicale d'un mot
Ex: ecrivions --> ecriv

**PorterStemmer** a √©t√© construit pour la langue anglaise et ne fonctionne pas tr√®s bien avec la langue francaise, il aura souvent du mal avec certains verbe

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stem_vocab = [stemmer.stem(word) for word in vocab]
print(stem_vocab[:50])

stem_vocab = set(stem_vocab)
print(len(stem_vocab))
```

**SnowballStemmer** a quand a lui √©t√© construit pour plusieurs langues dont le francais, il parvient donc mieux de l'utiliser lorsque les textes sont en francais

```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer('french')
stem_vocab = [stemmer.stem(word) for word in vocab]
print(stem_vocab[:50])

stem_vocab = set(stem_vocab)
print(len(stem_vocab))
```

# Lemmatization
Le fait de revenir a la forme canonique des mots
Ex: ecrivions --> ecrire

```python

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatizer_vocab = [lemmatizer.lemmatize(word) for word in vocab]
print(lemmatizer_vocab[:50])

lemmatizer_vocab = set(lemmatizer_vocab)
print(len(lemmatizer_vocab))
```

**Spacy** est une librairie permettant d'utiliser des models de ML & DL sur python, ce qui nous permet d'effectuer une lemmatization plus performante et puissante.

```python
# !for model in fr_core_news_md fr_core_news_sm fr_core_news_lg en_core_web_sm en_core_web_lg; do python -m spacy download $model; done 
# !python -m spacy validate
from nltk.corpus import stopwords
import spacy

nlp = spacy.load('fr_core_news_md')
vocab = set(
    doc for _, review in df_fr['review_body'].items() for doc in nlp(review)
)
print(len(vocab))
print(vocab)

vocab = vocab - set(stopwords.words('french'))
vocab = vocab - set(stopwords.words('english'))
print(len(vocab))
```

```python
!python -m pip install charset_normalizer 

def decode_charset(raw_bytes, suggested_encoding: str = None, decode_type: str='ignore'):
    """Fonction pour detecter le type d'encodage d'un test. 
    C'est ma version peut etre pas la meilleure mais jme suis inspir√© du prof"""
    from charset_normalizer import from_bytes as charset_from_bytes
    if suggested_encoding is not None: encoding = suggested_encoding; decode_type = 'strict'
    else: encoding = charset_from_bytes(raw_bytes).best().encoding

    if raw_bytes is None or encoding is None: return None, None

    try: decoded = raw_bytes.decode(encoding, decode_type)
    except(BaseException, Exception, ArithmeticError, BufferError, LookupError):
        encoding = charset_from_bytes(raw_bytes).best().encoding
        decoded = raw_bytes.decode(encoding, 'ignore')
    return decoded, encoding
```
